```python
#import basic libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#show all inbuilt datasets in sns
sns.get_dataset_names()
```

```python
#load dataset
data=sns.load_dataset('tips')
```

```python
#Show the top 5 rows of the dataset
data.head()
```

```python
#Scatterplot
sns.scatterplot(x="total_bill",y="total_bill",data=data)
```

```python
#Line plot
sns.lineplot(x="size",y="tip",data=data)
```

```python
#Boxplot
sns.boxplot(x="day",y="total_bill",data=data)
```

```python
#ViolinPlot
sns.violinplot(x="day",y="total_bill",data=data)
```

```python
#Barplot
sns.barplot(x="day",y="total_bill",data=data)
```

```python
#Barplot without errorbar
sns.barplot(x="day",y="total_bill",data=data, errorbar=None)
```

```python
#countplot
sns.countplot(x="day",data=data)
```

```python
#Swarm Plot

sns.swarmplot(x="day",y="tip",data=data)
```

```python
#histogram
sns.histplot(x="tip",data=data, kde=True)
```

```python
#KDE Plot
sns.kdeplot(x="tip",data=data)
```

```python
#heatmap

sns.heatmap(data.select_dtypes(include='number').corr(),annot=True,cmap="viridis")
```

```python
# converting the categorical values to
from sklearn.preprocessing import LabelEncoder
data_encoded=data.copy() #Making a copy of the original data
for col in data_encoded.columns:
  if data_encoded[col].dtype=='object' or str(data_encoded[col].dtype)=='category':
    data_encoded[col]=LabelEncoder().fit_transform(data_encoded[col])
sns.heatmap(data_encoded.corr(),annot=True,cmap="viridis")
```

```python
#Pair plot
sns.pairplot(data)
```
