```python
#Plot a circle
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
circle = plt.Circle((5, 5), 2, color='orange',fill=False,linewidth=2)
ax.add_artist(circle)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')
ax.legend()

plt.show()
```

##SymPy Library

```python
import sympy as sp
# Define symbolic variables

x, y = sp.symbols('x y')
expr = x**2 + 2*x*y + y**2
print(expr) #print a plain text version
expr
```

```python
from sympy import *
x, y = sp.symbols('x y')
expr = (x+2)*(x+3)
print("Expanded", expand(expr)) #expand
print("Foctored", factor(expand(expr))) #factor
expr2 = 5*x**2 - 4*x**2
print("Simplified", simplify(expr2)) #simplify
simplify(expr2)
```

```python
#Differentiation and Integration
x=symbols('x')
f = sin(x) + x**2
integrate(f,x)
```

```python
from sympy import *
x, y = sp.symbols('x y')
eq1 = Eq(x**2 -4, 0)
solve(eq1,x)

# System of equations
eq2=Eq(x + y, 5)
eq3=Eq(2*x - y, 1)
solve((eq2, eq3), (x, y))
```

```python
A = Matrix([[2,3],[1,4]])
print("Matrix:",A)
print("Determinant:",A.det())
print("Inverse:",A.inv())
print("Transpose:",A.T)
print("EigenValues",A.eigenvals())
print("EigenVectors",A.eigenvects())
```

```python
#Matrix Multiplication and Rank

A = Matrix([[1, 2], [3, 4]])
B = Matrix([[5, 6], [7, 8]])
C = A * B
print(C)
print("Rank",A.rank())
```

```python
from sympy.plotting import plot

f = sin(x) + cos(2*x)
plot(f, (x, -10, 10), title="y = sin(x) + cos(2x)")
```

```python
from sympy import Eq
from sympy.plotting import plot_implicit
y = symbols('y')

circle = Eq(x**2 + y**2, 9)
plot_implicit(circle, (x, -5, 5), (y, -5, 5), title='x + y = 9')
```

```python
from sympy.plotting import plot3d

f = sin(x)*cos(y)
plot3d(f, (x, -5, 5), (y, -5, 5), title='3D Surface: sin(x)cos(y)')
```

```python
from sympy import Piecewise

f = Piecewise((x**2, x < 0), (x, True))
plot(f, (x, -5, 5), title='Piecewise Function')
```

```python
from sympy import exp

f1 = sin(x)
f2 = cos(x)
f3 = exp(-x**2)

plot(f1, f2, f3, (x, -5, 5), legend=True, title='Multiple SymPy Functions')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import lambdify

# Symbolic expression
x = symbols('x')
expr = sin(x) + 0.5*cos(2*x)

# Convert to numerical function
f = lambdify(x, expr, 'numpy')

# Generate data and plot
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, color='purple', lw=2)
plt.title("y = sin(x) + 0.5cos(2x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
