```python
# Import libraries
import matplotlib.pyplot as plt
```

##Line Plot

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#plot
plt.plot(x,y)
plt.title("Basic Linear Line Plot")
plt.xlabel("X-Axis name")
plt.ylabel("Y-Axis name")
plt.show()
```

```python
#Customization of line plot

x=[1,2,3,4,5]
y=[2,4,6,8,10]

#plot
plt.plot(x,y, color="red", linestyle="-.", marker="o", markersize=6, linewidth=2)
plt.title("Basic Linear Line Plot")
plt.xlabel("X-Axis name")
plt.ylabel("Y-Axis name")
plt.show()
```

```python
#Multiple lines in a single plot

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

#plot
plt.plot(x,y1, color="red", linestyle="-.", marker="o", markersize=6, linewidth=2)
plt.plot(x,y2, color="blue", linestyle="--", marker="o", markersize=6, linewidth=2)
plt.title("Multiple Line Plot")
plt.xlabel("X-Axis name")
plt.ylabel("Y-Axis name")
plt.show()
```

```python
#Grid Lines and ticks

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

#plot
plt.plot(x,y1, color="red", linestyle="-.", marker="o", markersize=6, linewidth=2)
plt.plot(x,y2, color="blue", linestyle="--", marker="o", markersize=6, linewidth=2)
plt.title("Multiple Line Plot")
plt.xlabel("X-Axis name")
plt.ylabel("Y-Axis name")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,5,10,15,20,25])
plt.grid(True,linestyle="--",alpha=0.6)
plt.show()
```

```python
#Subplot

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

#subplot
plt.subplot(1,3,1)
plt.plot(x,y1, color="red", linestyle="-.", marker="o", markersize=6, linewidth=2)
plt.title("First subplot")

plt.subplot(1,3,2)
plt.plot(x,y2, color="red", linestyle="--", marker="o", markersize=6, linewidth=2)
plt.title("second subplot")

plt.subplot(1,3,3)
plt.plot(x,y2, color="red", linestyle="--", marker="o", markersize=6, linewidth=2)
plt.title("third subplot")
# plt.xlabel("X-Axis name")
# plt.ylabel("Y-Axis name")
# plt.grid(True,linestyle="--",alpha=0.6)
plt.show()
```

```python
#Style

plt.style.use("ggplot")

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

#plot
plt.plot(x,y1, color="red", linestyle="-.", marker="o", markersize=6, linewidth=2)
plt.plot(x,y2, color="blue", linestyle="--", marker="o", markersize=6, linewidth=2)
plt.title("Multiple Line Plot")
plt.xlabel("X-Axis name")
plt.ylabel("Y-Axis name")
plt.grid(True,linestyle="--",alpha=0.6)
plt.show()
```

```python
#Annotate with arrowprop


x=[1,2,3,4,5]
y1=[2,4,6,8,10]

#plot
plt.plot(x,y1, color="red", linestyle="-.", marker="o", markersize=6, linewidth=2)
plt.title("Line Plot")
plt.xlabel("X-Axis name")
plt.ylabel("Y-Axis name")

plt.annotate("4th point",xy=(4,8),xytext=(3,8),arrowprops=dict(facecolor="black"))
plt.show()
```

```python
#Legend

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

#plot
plt.plot(x,y1, color="red", label= "Linear", linestyle="-.", marker="o", markersize=6, linewidth=2)
plt.plot(x,y2, color="blue", label="Non-Linear", linestyle="--", marker="o", markersize=6, linewidth=2)
plt.title("Multiple Line Plot")
plt.xlabel("X-Axis name")
plt.ylabel("Y-Axis name")
plt.legend(loc="upper left")
plt.show()
```

##BAR Graph

```python
categories=["A","B","C","D"]
values=[3,5,7,9]

plt.bar(categories,values, color="orange", width=0.5,edgecolor="black")
plt.title("Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Horizontal Bar Graph

categories=["A","B","C","D"]
values=[3,5,7,9]

plt.barh(categories,values, color="orange",edgecolor="black")
plt.title("Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Annotations
categories=["A","B","C","D"]
values=[3,9,5,7]


plt.bar(categories,values, color="orange", width=0.5,edgecolor="black")

#Annotations
Labels= ["Low", "High","Medium","Above Avg"]
for i,label in enumerate(Labels):
  plt.text(i, values[i]+0.3, label, ha="center")

plt.annotate("Medium bar",xy=("C",5),xytext=("C",6),arrowprops=dict(facecolor="black"))

plt.title("Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
