```python
np.random.seed(1)
data=np.random.randn(1000)
print(data)
```

```python
#Histogram

plt.hist(data,bins=20,color="skyblue",edgecolor="black")
plt.title("Histogram example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(data,bins=20,color="orange", alpha=0.7, edgecolor="black")
plt.title("Histogram example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True,linestyle="--",alpha=0.3)
plt.show()
```

```python
# Scatter Plot

x=np.random.rand(50)*10
y= 2*x+np.random.randn(50)*2
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#customized scatter plot
plt.scatter(x,y,c="red",s=100,alpha=0.5,edgecolor="black")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True,linestyle="--",alpha=0.3)
plt.show()
```

```python
#Box Plot

np.random.seed(30)
data1=np.random.normal(50,10,200) #mean=50, std=10
data2=np.random.normal(60,15,200) #mean=60, std=15

plt.boxplot([data1,data2],labels=["Group 1","Group 2"])
plt.title("Box Plot")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

```python
plt.boxplot([data1,data2],labels=["Group 1","Group 2"],patch_artist=True,boxprops=dict(facecolor="lightblue"),medianprops=dict(color="red",linewidth=2),whiskerprops=dict(color="green",linewidth=1),capprops=dict(color="orange",linewidth=1))
plt.title("Box Plot")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

```python
#Lolipop Graph
categories=["a","b","c","d","e"]
value=np.array([10,25,15,30,20])

plt.stem(categories,value,basefmt=" ")
plt.title("Stem Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
