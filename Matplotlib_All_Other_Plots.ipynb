```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Data
sizes = [30, 20, 25, 25]
labels = ["A", "B", "C", "D"]
colors = ["skyblue", "lightgreen", "salmon", "gold"]

# Basic pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()


```

```python
# Donut Chart
plt.pie(sizes, labels=labels, wedgeprops=dict(width=0.5))
plt.title("Donut Chart")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.fill_between(x, y1, color="skyblue", alpha=0.5, label="sin(x)")
plt.fill_between(x, y2, color="salmon", alpha=0.5, label="cos(x)")
plt.title("Area Plot")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data: sales over months
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug"]
sales = [120, 180, 250, 300, 400, 450, 500, 600]

# Create area plot
plt.fill_between(months, sales, color="skyblue", alpha=0.5, label="Sales Growth")
plt.plot(months, sales, color="blue", marker="o")  # line on top for clarity

# Add labels and title
plt.title("Monthly Sales Growth")
plt.xlabel("Months")
plt.ylabel("Sales (in units)")
plt.legend()

# Show the plot
plt.show()
```

```python
x = np.arange(0, 10, 1)
y = np.random.randint(1, 10, size=10)

plt.step(x, y, where="mid", label="Step Plot", color="purple")
plt.scatter(x, y, color="red")  # markers to show actual points
plt.title("Step Plot Example")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = [10, 15, 12, 18, 14]
y_err = [1, 2, 1.5, 2.5, 1]  # uncertainty

plt.errorbar(x, y, yerr=y_err, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bar Example")
plt.xlabel("Month")
plt.ylabel("Sales (units)")
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)

plt.stem(x, y, basefmt="gray", linefmt="b-", markerfmt="ro")
plt.title("Stem Plot")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Random 3D data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x, y, z, c=z, cmap="viridis", s=60)
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
days = np.arange(1, 6)
men = [1, 2, 3, 4, 5]
women = [1, 1, 2, 3, 4]
children = [2, 1, 2, 1, 2]

plt.stackplot(days, men, women, children, labels=["Men", "Women", "Children"], alpha=0.6)
plt.legend(loc="upper left")
plt.title("Stackplot Example")
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

plt.contour(X, Y, Z, levels=15, cmap="viridis")
plt.title("Contour Plot")
plt.colorbar(label="Function Value")
plt.show()
```

```python
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
U = -Y   # x-direction vector
V = X    # y-direction vector

plt.quiver(X, Y, U, V, color="red")
plt.title("Quiver Plot (Vector Field)")
plt.show()
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hexbin(x, y, gridsize=30, cmap="Blues")
plt.colorbar(label="Count in bin")
plt.title("Hexbin Plot")
plt.show()
```

```python
x = np.random.rand(30)
y = np.random.rand(30)
sizes = np.random.randint(100, 1000, 30)

plt.scatter(x, y, s=sizes, alpha=0.5, c=sizes, cmap="viridis")
plt.title("Bubble Plot")
plt.colorbar(label="Bubble Size")
plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, 100)
r = 1 + np.sin(5*theta)

plt.polar(theta, r, color="purple")
plt.title("Polar Plot Example")
plt.show()
```
